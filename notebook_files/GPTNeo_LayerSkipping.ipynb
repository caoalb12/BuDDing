```python
import random, time

prompts = [
    # Common phrases and expressions (1-25)
    "Once upon a time, there was a",
    "I woke up this morning and decided to",
    "The weather outside is looking quite",
    "After a long day at work, I",
    "My favorite thing to do on weekends is",
    "When I was younger, I used to",
    "The best part about summer is",
    "If I could travel anywhere, I would go",
    "Yesterday I went to the store and bought",
    "The most important thing in life is",
    "During the winter months, people often",
    "At the end of the day, what matters most",
    "Walking down the street, I noticed",
    "In the middle of the night, I heard",
    "The first thing I do every morning is",
    "Looking out the window, I can see",
    "Before going to bed, I always",
    "On my way home, I stopped to",
    "The sound of rain makes me feel",
    "When the sun sets, the sky turns",
    "Inside the house, everything was",
    "Next week I'm planning to",
    "The smell of fresh coffee reminds me of",
    "As I opened the door, I saw",
    "Standing at the edge of the cliff, I",

    # Sentence patterns and structures (26-50)
    "She picked up the phone and",
    "He walked into the room carrying a",
    "They decided to meet at the",
    "The children were playing in the",
    "Everyone agreed that the meeting was",
    "The old man sat on the bench and",
    "With a smile on her face, she",
    "The dog ran across the yard chasing a",
    "By the time we arrived, the show had",
    "Without saying a word, he turned and",
    "The book on the table was about",
    "After careful consideration, they chose to",
    "The train was scheduled to arrive at",
    "Through the window I could hear",
    "Beneath the surface of the water lay",
    "The letter in the mailbox was from",
    "Despite the cold weather, we decided to",
    "Around the corner stood a large",
    "Behind the curtain was hidden a",
    "Among the crowd, there was one person who",
    "Between the two buildings was a narrow",
    "Above the clouds, the sun was",
    "Below the ground, miners were working to",
    "Across the river, we could see the",
    "Throughout history, people have always",

    # Narrative and storytelling (51-75)
    "The door slowly creaked open to reveal",
    "In a small village far away, there lived",
    "The journey began early in the morning when",
    "Nobody expected what would happen next when",
    "The secret had been hidden for years until",
    "As the clock struck midnight, something strange",
    "The treasure map showed a path leading to",
    "With trembling hands, she opened the mysterious",
    "The mysterious stranger appeared out of nowhere and",
    "In the distance, a figure was approaching",
    "The old photograph showed a place that",
    "According to legend, the ancient castle was",
    "Years ago, in this very spot, there",
    "The message written on the wall said",
    "Deep in the forest, explorers discovered",
    "The voice on the phone sounded",
    "When morning came, they realized that",
    "The footprints in the snow led to",
    "At the bottom of the box was",
    "The painting depicted a scene from",
    "Through the keyhole, I could barely see",
    "The rumor spreading through town was that",
    "In her pocket, she found a",
    "The shadow on the wall looked like",
    "Before sunrise, the fishermen would",

    # Descriptive and sensory (76-100)
    "The taste of the soup was surprisingly",
    "The texture of the fabric felt",
    "Colors in the garden included red, yellow, and",
    "The sound of thunder echoed across the",
    "Bright lights illuminated the entire",
    "The warmth of the fire made everyone",
    "A gentle breeze carried the scent of",
    "The rough surface of the wall was covered in",
    "Soft music played quietly in the background as",
    "The bitter cold wind blew through the",
    "Sweet aromas filled the kitchen while",
    "Heavy rain pounded against the",
    "The smooth glass reflected the image of",
    "Loud noises came from the direction of",
    "A thick fog rolled in from the",
    "The bright moonlight cast shadows on the",
    "Dark clouds gathered overhead, threatening",
    "The cool water felt refreshing after",
    "A faint whisper could be heard coming from",
    "The delicious meal consisted of",
    "Sharp objects were carefully placed in the",
    "The peaceful atmosphere was suddenly interrupted by",
    "A strange feeling came over me when",
    "The narrow path wound its way through",
    "Towering mountains rose high above the",

    # Conversational and relatable (101-128)
    "I remember thinking to myself that",
    "It's funny how things sometimes",
    "People always say that you should",
    "The problem with this approach is that",
    "What really matters at the end of",
    "Sometimes it feels like the world is",
    "The reason this happens is because",
    "If you think about it carefully, you'll notice",
    "It's important to remember that not everyone",
    "The difference between these two options is",
    "I've always wondered why people tend to",
    "The main advantage of doing it this way",
    "Looking back now, I realize that",
    "It takes a lot of effort to",
    "One thing I've learned over time is",
    "The challenge we face today is",
    "Most people don't understand how",
    "It's becoming increasingly clear that",
    "The simple truth is that sometimes",
    "When you compare the two, you'll see",
    "There's something special about the way",
    "It makes sense when you consider that",
    "The key to success is knowing when",
    "After all this time, I finally understand",
    "The real question we should be asking",
    "What surprised me the most was",
    "In my experience, the best approach is",
    "At first glance, it might seem like"
]

skip_layers = []
num_layers = 32

for _ in prompts:
    layers_to_skip = random.sample(range(4, num_layers - 4), 4)
    skip_layers.append(layers_to_skip)

print(skip_layers)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import torch.autograd.profiler as profiler

device = "cuda" if torch.cuda.is_available() else "cpu"

model_name = "EleutherAI/gpt-neo-2.7B"

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16).to(device)

model.eval() # turn off training mode

tokenizer_output = tokenizer(prompts, return_tensors="pt", padding=True)
input_ids = tokenizer_output.input_ids.to(device)
attention_mask = tokenizer_output.attention_mask.to(device)

generated_ids = input_ids.clone() # will add new tokens to this

batch_size = len(prompts)

max_tokens = 10
top_k = 5

with profiler.profile(record_shapes=True, profile_memory=True, use_cuda=True) as prof:
  net_start = time.time()

  for iter in range(max_tokens):
      hidden_states = model.transformer.wte(generated_ids) # token embeddings

      position_ids = torch.arange(
          generated_ids.shape[1],
          device=device
      ).unsqueeze(0)
      position_embeds = model.transformer.wpe(position_ids)

      hidden_states = hidden_states + position_embeds # add position embeddings

      if iter > 0:
        attention_mask = torch.cat([attention_mask, torch.ones((batch_size, 1), device=device)], dim=-1)

      seq_len = attention_mask.shape[1]

      causal_mask = torch.tril(torch.ones((seq_len, seq_len), device=device)).view(1, 1, seq_len, seq_len) # earlier tokens can't attend to future tokens
      attention_mask_4d = attention_mask.view(batch_size, 1, 1, seq_len) # ensure no tokens attend to <PAD> tokens

      combined_mask = causal_mask * attention_mask_4d # combine both masks
      combined_mask = (1.0 - combined_mask) * torch.finfo(torch.float16).min

      for layer_idx, layer in enumerate(model.transformer.h):
          start_time = time.time()

          active_indices = [i for i in range(batch_size) if layer_idx not in skip_layers[i]]

          if not active_indices:
              continue

          mini_batch = hidden_states[active_indices]
          mini_batch_mask = combined_mask[active_indices]

          mini_batch = layer(mini_batch, attention_mask=mini_batch_mask)[0]
          hidden_states[active_indices] = mini_batch

          end_time = time.time()
          if iter == 0: print(f"Layer {layer_idx}: Prompts = {len(active_indices)}, Forward Pass Time: {end_time - start_time:.4} sec")

      hidden_states = model.transformer.ln_f(hidden_states) # layer norm

      logits = model.lm_head(hidden_states) # project to vocabulary

      probs = torch.softmax(logits[:, -1, :], dim=-1)
      topk_probs, topk_indices = torch.topk(probs, top_k, dim=-1)

      next_token_idx = torch.multinomial(topk_probs, num_samples=1)
      next_token_id = topk_indices.gather(-1, next_token_idx)

      generated_ids = torch.cat([generated_ids, next_token_id], dim=-1)

  net_end = time.time()

  print(f"Net Time: {net_end - net_start:.4} sec")

print(prof.key_averages().table(sort_by="cuda_time_total", row_limit=40))

for i, ids in enumerate(generated_ids):
    text = tokenizer.decode(ids, skip_special_tokens=True).replace("\n", " ")
    print(f"Prompt {i}: {text}")
```
