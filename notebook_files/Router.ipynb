```python
from datasets import load_dataset, concatenate_datasets

sample_size = 100  # or whatever you wan
combined_dataset = []
```

```python
wino = load_dataset("winogrande", "winogrande_xl", split="train").select(range(sample_size))

def winogrande_prompt(sentence, option):
    """
    Insert the option into the blank (_) in the Winogrande sentence.
    """
    return sentence.replace("_", option)

for ex in wino:
    p1 = winogrande_prompt(ex["sentence"], ex["option1"])
    p2 = winogrande_prompt(ex["sentence"], ex["option2"])
    combined_dataset.append(p1)
    combined_dataset.append(p2)
```

```python
hella = load_dataset("hellaswag", split="train").select(range(sample_size))

def hellaswag_prompt(context, ending):
    """
    Combine context + ending for HellaSwag.
    """
    # You can customize how you join context + ending; simplest:
    return context + " " + ending

for ex in hella:
    context = ex["context"]
    endings = ex["endings"]  # list of 4 endings

    for end in endings:
      combined_dataset.append(hellaswag_prompt(context, end))
```
